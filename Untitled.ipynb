```python
import pandas as pd
import numpy as np 
import matplotlib as plt
import seaborn as sns
from scipy.stats import norm
import statistics
import math
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("stock csv/ADEL Historical Data.csv")
df["Change %"] = (
    df["Change %"]
        .astype(str)
        .str.replace("%", "", regex=False)
)

df["Change %"] = pd.to_numeric(df["Change %"], errors="coerce")
df["Date"] = pd.to_datetime(df["Date"], errors="coerce" )
df.head()
```

```python
def text_to_num(text):
    d = {'K': 1000, 'M': 1000000, 'B': 1000000000, 'T': 1000000000000}
    if not isinstance(text, str):
        return text
    
    # Extract numerical part and suffix
    text = text.upper()
    multiplier = 1
    for suffix, value in d.items():
        if suffix in text:
            multiplier = value
            text = text.replace(suffix, '')
            break
            
    return float(text) * multiplier

df["Vol."] = df["Vol."].apply(text_to_num)
df["Vol."] =  pd.to_numeric(df["Vol."],errors="coerce")
```

```python
df  = df.sort_values("Date",ascending=False)
df.dropna()
```

```python
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

X = df[['Price']].values
y = df['Open'].values

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=0
)

model = LinearRegression()
model.fit(X_train, y_train)

print("Intercept:", model.intercept_)
print("Coefficient:", model.coef_[0])

y_pred = model.predict(X_test)

# Visualization
plt.scatter(X_train, y_train, color='green')
plt.plot(X_test, y_pred, color='black', linewidth=3)
plt.title('Price vs Open')
plt.xlabel('Price')
plt.ylabel('Open')
plt.show()
```

```python
import statistics

df = df.dropna()
df['Date'] = pd.to_datetime(df['Date'])
price_mean = {}
price_stdev = {}
for x in range(0,22):
    yer = 2000 + x
    filtered_df = df[df['Date'] < f'{yer}-01-01'].copy()
    

    print(f"------------------- {yer} -------------------")

    if len(filtered_df) >= 2:
        price_mean[yer] = round(filtered_df["Price"].mean(), 2)
        price_stdev[yer] =  round(statistics.stdev(filtered_df["Price"]), 2)
        print(" Price mean is", round(filtered_df["Price"].mean(), 2))
        print(" Price median is", round(filtered_df["Price"].median(), 2))
        print(" Price stdev is", round(statistics.stdev(filtered_df["Price"]), 2))
        
        print(" High mean is", round(filtered_df["High"].mean(), 2))
        print(" High median is", round(filtered_df["High"].median(), 2))
        print(" High stdev is", round(statistics.stdev(filtered_df["High"]), 2))

        print(" Low mean is", round(filtered_df["Low"].mean(), 2))
        print(" Low median is", round(filtered_df["Low"].median(), 2))
        print(" Low stdev is", round(statistics.stdev(filtered_df["Low"]), 2))

        print(" correlation is Price and Open", round(statistics.correlation(filtered_df["Price"],filtered_df["Open"]), 3))
        print(" correlation is High and Low", round(statistics.correlation(filtered_df["High"],filtered_df["Low"]), 3))
    else:
        print("Not enough data for statistics")
```

```python
# Convert to DataFrame
df = pd.DataFrame({
    "Year": list(price_mean.keys()),
    "Mean": list(price_mean.values()),
    "Volatility": list(price_stdev.values())
}).sort_values("Year").reset_index(drop=True)

# ----------------------------
# 1 LOWEST YEAR
# ----------------------------

lowest_row = df.loc[df["Mean"].idxmin()]
lowest_year = lowest_row["Year"]
lowest_value = lowest_row["Mean"]

# ----------------------------
# 3️YEAR-OVER-YEAR GROWTH
# ----------------------------

df["YoY_Growth_%"] = df["Mean"].pct_change() * 100

biggest_growth = df.loc[df["YoY_Growth_%"].idxmax()]
biggest_drop = df.loc[df["YoY_Growth_%"].idxmin()]

# ----------------------------
# GROWTH START (Trend Reversal)
# ----------------------------

growth_start = None
for i in range(1, len(df)-1):
    if df.loc[i, "Mean"] > df.loc[i-1, "Mean"] and \
       df.loc[i+1, "Mean"] > df.loc[i, "Mean"]:
        growth_start = df.loc[i, "Year"]
        break

# ----------------------------
# SUSTAINED UPTREND (3 years rising)
# ----------------------------

sustained_trend = None
for i in range(2, len(df)):
    if (df.loc[i, "Mean"] > df.loc[i-1, "Mean"] and
        df.loc[i-1, "Mean"] > df.loc[i-2, "Mean"]):
        sustained_trend = df.loc[i-2, "Year"]
        break

# ----------------------------
# MOST VOLATILE YEAR
# ----------------------------

most_volatile = df.loc[df["Volatility"].idxmax()]

# ----------------------------
# VOLATILITY REGIME CHANGE
# ----------------------------

regime_change = None
for i in range(1, len(df)):
    if df.loc[i, "Volatility"] > 3 * df.loc[i-1, "Volatility"]:
        regime_change = df.loc[i, "Year"]
        break

# ----------------------------
# CAGR
# ----------------------------

start_value = df.iloc[0]["Mean"]
end_value = df.iloc[-1]["Mean"]
years_count = len(df) - 1

cagr = ((end_value / start_value) ** (1 / years_count) - 1) * 100

# ----------------------------
#  LONGEST BULL & BEAR RUN
# ----------------------------

bull_run = bear_run = 0
max_bull = max_bear = 0

for i in range(1, len(df)):
    if df.loc[i, "Mean"] > df.loc[i-1, "Mean"]:
        bull_run += 1
        bear_run = 0
    elif df.loc[i, "Mean"] < df.loc[i-1, "Mean"]:
        bear_run += 1
        bull_run = 0

    max_bull = max(max_bull, bull_run)
    max_bear = max(max_bear, bear_run)

# ----------------------------
# PRINT RESULTS
# ----------------------------

print("===== MARKET ANALYSIS RESULTS =====")
print(f"Lowest Year: {lowest_year} (Mean={lowest_value})")
print(f"Growth Started: {growth_start}")
print(f"Sustained Uptrend Started: {sustained_trend}")
print(f"Biggest Growth Year: {int(biggest_growth['Year'])} ({biggest_growth['YoY_Growth_%']:.2f}%)")
print(f"Biggest Drop Year: {int(biggest_drop['Year'])} ({biggest_drop['YoY_Growth_%']:.2f}%)")
print(f"Most Volatile Year: {int(most_volatile['Year'])}")
print(f"Volatility Regime Change: {regime_change}")
print(f"CAGR (2001-2021): {cagr:.2f}%")
print(f"Longest Bull Run: {max_bull} years")
print(f"Longest Bear Run: {max_bear} years")

# ----------------------------
# PLOTS
# ----------------------------
```
